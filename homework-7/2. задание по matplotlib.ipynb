```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

## Exercise 1  
1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!

```python
x = np.linspace(-1, 1, 250)
y = x ** 2 * np.sin(1 / x ** 2) + x
fig, ax = plt.subplots()

ax.plot(x, y, color = 'green')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
```

2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)

```python
x = np.linspace(5, 25, 250)
y_1 = 1 / (1 + x ** 2)
y_2 = 1 / x ** 2
y = 1 - y_1 / y_2
fig, ax = plt.subplots()

ax.semilogy(x, y, color = 'blue')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('Относительная разница', fontsize=12)
```

3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>! Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.

```python
from scipy.special import laguerre

x = np.linspace(-1, 5, 50)
fig, ax = plt.subplots()

ax.plot(x, laguerre(1)(x), linestyle = '', marker = '1', color = 'red', label=rf'$L_{1}$')
ax.plot(x, laguerre(2)(x), linestyle = '', marker = '+', color = 'blue', label=rf'$L_{2}$')
ax.plot(x, laguerre(3)(x), linestyle = '--', color = 'black', label=rf'$L_{3}$')
ax.plot(x, laguerre(4)(x), linestyle = '-', marker = '', color = 'green', label=rf'$L_{4}$')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('Многочлены Лагерра', fontsize=12)
plt.legend()
```

```python
from scipy.special import factorial

x = np.linspace(1, 5, 5)
y = factorial(x)
fig, ax = plt.subplots()

ax.bar(x, y, color = 'pink')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('x!', fontsize=12)
ax.grid()
```

## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
x = np.linspace(-6 * np.pi, 6 * np.pi, 100)
y = np.sin(x) / x
fig, ax = plt.subplots(figsize=(16.5,5))

ax.plot(x, y, marker='*', linestyle='', markersize=18, markerfacecolor='yellow', markeredgecolor='black')
ax.set_xticks(np.arange(-6 * np.pi, 6 * np.pi + np.pi / 2, np.pi/2))
i = -6
tickl = [str(i) + r'$\pi$']
while i < 6:
    i += 1/2
    tickl.append(str(i) + r'$\pi$')
ax.set_xticklabels(tickl)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.grid()
```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.

```python
x = y = np.linspace(-np.pi, np.pi, 250)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y)
fig, ax = plt.subplots(figsize=(6, 5))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.pcolor(X, Y, Z, shading = 'auto', norm=norm, cmap='GnBu_r')

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)

cb = fig.colorbar(p, ax=ax)
cb.set_label('z', fontsize=12)
cb.set_ticks([-1, -0.5, 0, 0.5, 1])
```

2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
x = y = np.linspace(-3, 3, 250)
X, Y = np.meshgrid(x, y)
Z = np.exp(- X ** 2 - Y ** 2)
fig, ax = plt.subplots(figsize=(6, 5))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.pcolor(X, Y, Z, shading = 'auto',norm=norm, cmap='twilight_r')

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)

cb = fig.colorbar(p, ax=ax)
cb.set_label('z', fontsize=12)
cb.set_ticks([-1, -0.5, 0, 0.5, 1])
```
