13. Посчитайте две матрицы: $B = A^TA$ и $C = AA^T$. Что вы можете о них сказать? Какого они размера? В чём их сходство и отличие?

```python
A = np.array([[2, 7, 6.], 
              [0, -2, -1.], 
              [9, 6, 3.]])

B = A.T.dot(A)
C = A.dot(A.T)
print("Матрица B:", B, sep = "\n")
print("Матрица C:", C, sep = "\n")
```

Матрицы B и C имеют размер такой же, как исходная матрица A. Их сходство заключается в том, что в них одинаковое количество строк и столбцов (элементов), они обе получаются перемножением исходной матрицы A и ее транспонированной версии. Элементы матриц B и C отличаются друг от друга, т. к. перемножаются матрица A и транспонированная матрица A в разном порядке.

```python
def f(x):
    return np.cos(np.log(x)) / x

def pryamoug(f, a, b, n):
    x = np.linspace(a, b, n + 1)
    step = x[1] - x[0]
    f_v = np.vectorize(f)
    res = f_v(x)
    return np.sum(res) * step

def trapec(f, a, b, n):
    x = np.linspace(a, b, n + 1)
    step = x[1] - x[0]
    f_v = np.vectorize(f)
    res = f_v(x)
    N = len(res) - 1
    return step / 2 * (res[0] + 2 * np.sum(res[1:N]) + res[N])

def simps(f, a, b, n):
    x = np.linspace(a, b, n + 1)
    step = x[1] - x[0]
    f_v = np.vectorize(f)
    res = f_v(x)
    N = len(res) - 1
    return step / 3 * (res[0] + 2 * np.sum(res[2: N :2]) + 4 * np.sum(res[1: N :2]) + res[N])
```

```python
a = 1
b = np.exp(1)
n = 10
answ_pr = pryamoug(f, a, b, n)
answ_tr = trapec(f, a, b, n)
answ_sp = simps(f, a, b, n)
res = np.hstack([answ_pr, answ_tr, answ_sp])

while np.max(res) - np.min(res) >= 0.00001:
    n *= 10
    res[0] = pryamoug(f, a, b, n)
    res[1] = trapec(f, a, b, n)
    res[2] = simps(f, a, b, n)

print("Номер метода с наименьшим значением =", np.argmin(res) + 1)
print("Номер метода с наибольшим значением =", np.argmax(res) + 1)
print("Количество отрезков разбиения =", n)
print("Ширина шага =", (b - a) / n)
```

При сравнении получаемых приближенными методами ответов между собой получилось, что наименьший ответ был получен методом Симпсона, а наибольший - методом прямоугольников. Отличие в ответах будет при 5-м знаке после запятой при шаге примерно 1,7e-06 с разбиением на 1000000 шагов

```python
answer = np.tile(np.sin(1), 3)
razn = np.abs(answer - res)
print("Номер метода с наименьшим отклонением от точного ответа =", np.argmin(razn) + 1)
print("Номер метода с наибольшим отклонением от точного ответа =", np.argmax(razn) + 1)
```

При сравнении полученных приближенными методами при одинаковом количестве разбиений ответов с точным ответом получилось, что метод трапеций в данном случае дает наиболее приближенный к точному результат, в то время как метод прямоугольников - наименее близкий при одинаковом отличии между ответами в 5-м знаке после запятой.

```python
answer = np.sin(1)

n_pr = 10
while np.abs(answer - pryamoug(f, 1, np.exp(1), n_pr)) >= 0.00001:
    n_pr *= 10

n_tr = 10
while np.abs(answer - trapec(f, 1, np.exp(1), n_tr)) >= 0.00001:
    n_tr *= 10

n_sp = 10
while np.abs(answer - simps(f, 1, np.exp(1), n_sp)) >= 0.00001:
    n_sp *= 10

answ_pr = pryamoug(f, a, b, n_pr)
answ_tr = trapec(f, a, b, n_tr)
answ_sp = simps(f, a, b, n_sp)
res = np.hstack([answ_pr, answ_tr, answ_sp])

print("Количество шагов в методе прямоугольников =", n_pr)
print("Количество шагов в методе трапеций =", n_tr)
print("Количество шагов в методе Симпсона =", n_sp)
print("Ответ методом прямоугольников =", res[0])
print("Ответ методом трапеций =", res[1])
print("Ответ методом Симпсона =", res[2])

answer = np.tile(np.sin(1), 3)
razn = np.abs(answer - res)
print("Номер метода с наименьшим отклонением от точного ответа =", np.argmin(razn) + 1)
print("Номер метода с наибольшим отклонением от точного ответа =", np.argmax(razn) + 1)
```
